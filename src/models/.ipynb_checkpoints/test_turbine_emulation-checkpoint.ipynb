```python
# test_five_state_flow.py
from FEB_turbine_unit_U1 import make_unit, TurbineUnitConfig

u = make_unit(TurbineUnitConfig())
print("Initial:", u.status())  # SHUTDOWN, 0 MW

# Request 10 MW -> START (0 MW for 3 min)
u.request_output(10.0)
for m in [1, 2, 3]:
    p = u.step(1.0)
    print(f"t={m}m |", u.status())  # stay 0 MW until t=3m; after third minute, state moves to RAMP_UP

# RAMP_UP toward target with 1 MW/min
for m in range(5):
    p = u.step(1.0)
    print(f"ramp_up t={3+m+1}m |", u.status())  # will move into NORMAL when close to target (and within bounds)

# Energy look-ahead (does not mutate)
e15 = u.energy_in_minutes(15, mutate=False)
print("Projected energy next 15m:", e15, "MWh")

# Request stop -> RAMP_DOWN, then SHUTDOWN (0 MW during 5-min timer)
u.request_output(0.0)
for m in range(7):
    p = u.step(1.0)
    print(f"ramp_down/shutdown t=+{m+1}m |", u.status())
```
