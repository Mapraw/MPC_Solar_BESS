```python
# plot_from_module.py
from FEB_turbine_unit_U4 import HydroStorageConfig, HydroStorageUnit
import matplotlib.pyplot as plt

cfg = HydroStorageConfig(
    # max_discharge_mw=120, max_charge_mw=120,
    # min_discharge_mw=15, min_charge_mw=12,
    # ramp_mw_per_min=10,
    # capacity_mwh=300, initial_soe_mwh=150,
    # eta_charge=0.92, eta_discharge=0.90
)
u = HydroStorageUnit("U1", cfg)

# 4-hour schedule @ 1-min
schedule = []
for t in range(400):
    if t < 30:
        schedule.append(+7.0)
    elif t < 90:
        schedule.append(8.0)
    elif t < 150:
        schedule.append(5.0)
    else:
        schedule.append(+12.0)

time_min = list(range(401))
power_mw = [u.setpoint_mw]
soe_mwh = [u.soe_mwh]

for t in range(400):
    u.request_power(schedule[t])
    u.step(1.0)
    power_mw.append(u.setpoint_mw)
    soe_mwh.append(u.soe_mwh)

fig, ax1 = plt.subplots(figsize=(10,5))
ax1.plot(time_min, power_mw, label="Power (MW)", color="#1f77b4")
ax1.axhline(0, color="black", linewidth=0.8)
ax1.set_xlabel("Time (minutes)")
ax1.set_ylabel("Power to grid (+) / from grid () [MW]", color="#1f77b4")
ax1.tick_params(axis="y", labelcolor="#1f77b4")

ax2 = ax1.twinx()
ax2.plot(time_min, soe_mwh, label="SoE (MWh)", color="#ff7f0e")
ax2.set_ylabel("State of Energy (MWh)", color="#ff7f0e")
ax2.tick_params(axis="y", labelcolor="#ff7f0e")

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left")
ax1.set_title("Hydro Storage  Power & SoE (Policy A with Min Levels)")
plt.tight_layout()
plt.show()
```
